```python
import pandas as pd
from pandasql import sqldf
```

```python
filename_staff = "staff.csv"
filename_staffdetail = "staff_detail.csv"

table_staff = pd.read_csv(filename_staff)
table_staffdetail = pd.read_csv(filename_staffdetail)
```

```python
tablename = "table_staff"
sql = f"SELECT * FROM {tablename}"
df = sqldf(sql, globals())
df
```

```python
tablename = "table_staff"
sql = f"""
    SELECT `ID`, `DeptId`, `Gender` 
    FROM {tablename}
    WHERE `Salary` > 50000
    """
sqldf(sql, globals())
```

```python
sql_join = """
    SELECT *
    FROM table_staff 
    LEFT JOIN table_staffdetail 
    ON table_staff.ID = table_staffdetail.ID
"""
sqldf(sql_join, globals())
```

```python
"""
This will cause OperationalError:
    ambiguous column name: ID
"""

sql_join = """
    SELECT ID
    FROM table_staff 
    LEFT JOIN table_staffdetail 
    ON table_staff.ID = table_staffdetail.ID
"""
sqldf(sql_join, globals())
```

```python
sql_join = """
    SELECT table_staff.ID, table_staff.Name, table_staffdetail.Level, table_staffdetail.UpdateDate
    FROM table_staff
    LEFT JOIN table_staffdetail
    ON table_staff.ID = table_staffdetail.ID
"""
sqldf(sql_join, globals())
```

```python
sql_join = """
    SELECT st.ID, st.Name, std.Level, std.UpdateDate
    FROM table_staff st
    LEFT JOIN table_staffdetail std
    ON st.ID = std.ID
"""
sqldf(sql_join, globals())
```
