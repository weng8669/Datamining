## Primary keys name not the same

```python
filename_staff = "staff.csv"
filename_staffdetail = "staff_detail_different_col.csv"

df_staff = pd.read_csv(filename_staff)
df_staffdetail = pd.read_csv(filename_staffdetail)
```

```python
df_staff
```

```python
df_staffdetail
```

```python
pd.merge(df_staff, df_staffdetail)
```

```python
# Determine primary key and foreign key

pd.merge(df_staff, df_staffdetail, left_on="ID", right_on="EmpNum")
```

## Differentiate data source when two tables have the same column

```python
filename_staff = "staff_samecol.csv"
filename_staffdetail = "staff_detail_different_col.csv"

df_staff = pd.read_csv(filename_staff)
df_staffdetail = pd.read_csv(filename_staffdetail)
```

```python
df_staff
```

```python
df_staffdetail
```

```python
pd.merge(df_staff, df_staffdetail, left_on="ID", right_on="EmpNum", suffixes=["_left", "_right"])
```
