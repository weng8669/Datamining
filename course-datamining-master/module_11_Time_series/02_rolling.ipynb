```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(
    data=range(0, 10),
    columns=["testcol"]
)
df
```

```python
df.testcol.rolling(window=5).sum()
```

```python
df.testcol.rolling(window=5, min_periods=2).sum()
```

```python
"""
Customed defined function
For example -> Standard Deviation
"""

df2 = pd.DataFrame(
    np.random.randint(1, 10, 30),
    columns=["testcol"]
)
df2
```

```python
df2.testcol.rolling(window=5).std(ddof=0)
```

$\sigma = \sqrt{\frac{1}{n} \sum_{i=1}^n (x_i-\mu)^2}$

```python
df2.testcol.rolling(window=5).apply(lambda x: np.sqrt(np.power(x - x.mean(), 2).mean()))
```
