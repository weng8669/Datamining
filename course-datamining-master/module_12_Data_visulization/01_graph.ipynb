# $\frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{1}{2}(\frac{x-\mu}{\sigma})^2}$

```python
# 常態分佈

import numpy as np
import matplotlib.pyplot as plt

mean = 0; std = 1; variance = np.square(std)
x = np.arange(-5,5,.01)
f = np.exp(-np.square(x-mean)/2*variance)/(np.sqrt(2*np.pi*variance))

plt.plot(x,f)
plt.ylabel('gaussian distribution')
plt.show()
```

```python
plt.hist(np.random.normal(0, 1, 1000), 30, density=True)
plt.plot(x,f)
plt.ylabel('gaussian distribution')
plt.show()
```

# $\frac{1}{(1 + e^{-x})}$

```python
'''
sigmoid
'''

import numpy as np
import matplotlib.pyplot as plt

mean = 0; std = 1; variance = np.square(std)
x = np.arange(-5,5,.01)
f = 1/(1 + np.exp(-x))

plt.plot(x,f)
plt.ylabel('sigmoid')
plt.show()
```

# $\frac{e^{x}}{(1 + e^{x})}$

```python
'''
logit
'''

import numpy as np
import matplotlib.pyplot as plt

mean = 0; std = 1; variance = np.square(std)
x = np.arange(-5,5,.01)
f = np.exp(x) / (1 + np.exp(x))


plt.plot(x,f)
plt.ylabel('logit')
plt.show()
```
