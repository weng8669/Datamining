```python
import pandas as pd

filename = "ptt.csv"
df = pd.read_csv(filename)
```

```python
df
```

# 過濾器操作

```python
# 先將原資料進行備份
df2 = df.copy()
```

```python
'''
使用一般函數製作過濾器
'''

# 製作過濾器
def column_filter(s):
    return s.split(': ')[-1].strip()
```

```python
df['推數'].apply(column_filter)
```

```python
'''
上一步操作並不會直接改變 df 內容，
必須重新賦值才會改變 df 內容
'''

df['推數']
```

```python
'''
利用 for 迴圈對 DataFrame 取值，
可取得每個欄位名稱
'''

for c in df:
    print(c)
```

```python
for c in df:
    df[c] = df[c].apply(column_filter)
df
```

```python
'''
使用 lambda 作為過濾器
'''

df2['推數'].apply(lambda x: x.split(':')[-1].strip())
```

```python
for c in df2:
    df2[c] = df2[c].apply(lambda x: x.split(': ')[-1].strip())
df2
```
